# The basics of matplotlib
+ `.plot()` - временной ряд
+ `.plot()` - scatter plot
+ Разные стили маркеров
+ Отрисовка гистограмм
+ сохранить картинки

```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
fb = pd.read_csv('data/fb_stock_prices_2018.csv',
                index_col='date',
                parse_dates=True)
fb.head()
```

```python
plt.plot(fb.index, fb.open)
plt.show()
```

### Scatter plot с помощью обычного plot

```python
# Нарисуем scatter plot
# стиль маркеров укажем как or
# o - кружочки
# r - red
plt.plot('high', 'low', 'or', data=fb.head(20));
```

### Разные стили маркеров

```python
plt.plot('high', 'low', 'xm', data=fb.head(20));
```

```python
plt.plot('high', 'low', '.-.c', data=fb.head(20));
```

```python
plt.plot('high', 'low', 'xk', data=fb.head(20));
```

```python
plt.plot('high', 'low', 'x-b', data=fb.head(20));
```

### Отрисовка гистограмм

```python
quakes = pd.read_csv('data/earthquakes.csv')
quakes.head()
```

```python
# использую query для получения данных из датасета
quakes.query('magType == "ml"').mag.head()
```

```python
plt.hist(quakes.query('magType == "ml"').mag);
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,3))
# построим гистрограммы с разным количество бинов
axes
```

```python
fig
```

```python
x = quakes.query('magType == "ml"').mag
fig, axes = plt.subplots(1,2, figsize=(10,3))
for ax, bins in zip(axes, [7,35]):
    ax.hist(x, bins=bins)
    ax.set_title(f'bins params:{bins}')
```

```python
for ax, bins in zip(axes, [7,35]):
    print(bins)
```

```python
# сохранить картинки
x = quakes.query('magType == "ml"').mag
fig, axes = plt.subplots(1,2, figsize=(10,3))
for ax, bins in zip(axes, [7,35]):
    ax.hist(x, bins=bins)
    ax.set_title(f'bins params:{bins}')
    plt.savefig(f'fig {bins}.jpg')
```
