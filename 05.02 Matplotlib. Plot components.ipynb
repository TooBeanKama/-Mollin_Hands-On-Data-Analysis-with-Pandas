# Plot components

```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
# plt.figure создает объект БЕЗ осей
fig = plt.figure()
```

```python
# The plt.subplots() function creates a Figure object with Axes objects 
# for subplots in the arrangement specified.

# (row, columns)
fig, axes = plt.subplots(1,2)
```

```python
# мы можем добавить график внутрь другого графика
fig = plt.figure(figsize=(3,3))
outside = fig.add_axes([0.1, 0.1, 0.9, 0.9])
inside = fig.add_axes([0.7, 0.7, 0.25, 0.25])
```

```python
# создать отдельные графики разных размеров
fig = plt.figure(figsize=(8,8))
gs = fig.add_gridspec(3,3)
top_left = fig.add_subplot(gs[0,0])
mid_left = fig.add_subplot(gs[1,0])
top_right = fig.add_subplot(gs[:2,1:])
bottom = fig.add_subplot(gs[2,:])
```

```python
# сохранить пустые картинки
fig.savefig('empty.png')
```

+ `plt.<func>()` имеет доступт только к **последнему** Figure object
+ если мы сохраняем ссылки на **Figure objects**, мы можем работать с каждым из них, внезавимости от того, когда они были созданы
